### A.3 Ilustrační příklad - Hod tenisovým míčkem

Ve škole jsme se učili, jak popsat pohyb tenisového míčku vržené šikmo vzhůru:

\begin{align}
x(t) &= x_0 + v_0 \cos (\alpha) (t - t_0)\,, \\
y(t) &= y_0 + v_0 \sin (\alpha) (t - t_0) - \frac{1}{2} g (t - t_0)^2\,,
\end{align}

kde $[x_0, y_0]$ představuje počáteční polohu míčku, $v_0$ je jeho počáteční rychlost, $\alpha$ má význam úhlu hodu (vůči vodorovné přímce), $t_0$ odpovídá okamžiku hodu a $g$ je tíhové zrychlení.


parabola_3.png


Míček dopadne na zem v okamžiku $T$, když jeho výška bude nulová, tj. $y(T) = 0$. Abychom určili čas $T$, musíme vyřešit kvadratickou rovnici:

$$
-\frac{1}{2} g (T - t_0)^2 + v_0 \sin (\alpha) (T - t_0) + y_0 = 0\,.
$$

Ač má rovnice dvě řešení (reálné kořeny), jen jedno je fyzikální:

$$
T = t_0 + \frac{v_0 \sin (\alpha) + \sqrt{v_0^2 \sin^2 (\alpha) + 2 g y_0}}{g}\,.
$$

Známe-li $T$, můžeme spočítat $x$-ovou souřadnici míčku v okamžiku dopadu na zem:

$$
x(T) = x_0 + v_0 \cos (\alpha) (T - t_0)\,.
$$


Dále budeme uvažovat následující číselné hodnoty:
* $x_0 = 0~\mathrm{m}$, $y_0 = 1~\mathrm{m}$, $t_0 = 0~\mathrm{s}$, $v_0 = 25~\mathrm{m.s^{-1}}$, $\alpha = 45°$, $g = 10~\mathrm{m.s^{-2}}$

Pro okamžik dopadu (popř. doby celého letu) dostaneme:

$$
T = \frac{25 \times \frac{1}{\sqrt{2}} + \sqrt{25^2 \times \frac{1}{2} + 2 \times 20 \times 1}}{10}~\mathrm{s} \doteq 3,65~\mathrm{s}\,,
$$
z čeho dopočítáme vzdálenost $x(T)$:
$$
X(T) \doteq 25 \times \frac{1}{\sqrt{2}} \times 3,65~\mathrm{m} \doteq 64,5~\mathrm{m}\,.
$$

**Úkol**

Definuj funkce $x(t)$ a $y(t)$ a vykresli je pro zadané číselné hodnoty.

```python
# na začátek importujeme užitečné knihovny
import numpy as np
import matplotlib.pyplot as plt
```

```python
# počáteční podmínky a konstanty
x_0, y_0, t_0, v_0 = 0, 1, 0, 25
alpha = np.deg2rad(45) # nebo np.pi/4     funkce np.sin a np.cos očekává hodnotu úhlu v radiánech
g = 10
```

```python
# definujeme funkce

# x - ová souřadnice v čase t
def x_sour_teorie(t, t_0, x_0, v_0, alpha):
  return x_0 + v_0 * np.cos(alpha) * (t - t_0)

# y - ová souřadnice v čase t
def y_sour_teorie(t, t_0, x_0, v_0, alpha, g):
  return y_0 + v_0 * np.sin(alpha) * (t - t_0) - g * (t - t_0)**2 / 2

# T - okamžik dopadu
def T_cas_teorie(t_0, y_0, v_0, alpha, g):
  citatel    = v_0 * np.sin(alpha) + np.sqrt(v_0**2 * np.sin(alpha)**2 + 2 * g * y_0)
  jmenovatel = g
  return t_0 + citatel / jmenovatel
```

```python
# popřípade můžeme pro x a y -ové souřadnice v čase definovat jen jednu společnou funkci
def xy_sour_teorie(t, t_0, x_0, y_0, v_0, alpha, g):
  x = x_0 + v_0 * np.cos(alpha) * (t - t_0)
  y = y_0 + v_0 * np.sin(alpha) * (t - t_0) - g * (t - t_0)**2 / 2
  return x, y
```

```python
# pro ilustraci vypíšeme hodnoty v souřadnic v čase t = 1 s
t = 1
x_t = x_sour_teorie(t, t_0, x_0, v_0, alpha)
y_t = y_sour_teorie(t, t_0, x_0, v_0, alpha, g)

print("x(t = " + str(t) + " s) = " + str(x_t) + " m")   # pro zaokrouhlení: round(x_t,1)
print("y(t = " + str(t) + " s) = " + str(y_t) + " m")
```

```python
# popřípade ještě takto
t = 1
x_t, y_t = xy_sour_teorie(t, t_0, x_0, y_0, v_0, alpha, g)

print("x(t = " + str(t) + " s) = " + str(x_t) + " m")
print("y(t = " + str(t) + " s) = " + str(y_t) + " m")
```

```python
# určíme okamžik dopadu
T = T_cas_teorie(t_0, y_0, v_0, alpha, g)
print("Doba letu = " + str(round(T,2)) + " s")

# vytvoříme pole (array) reprezentující časové okamžiky, bude mít 100 komponent
# první okamžik je t_0 (okamžik hodu), poslední okamžik je T (okamžik dopadu)
t_pole = np.linspace(t_0, T, 100)

# pole souřadnice x - namísto jednoho času t (output by bylo jedno číslo) jsme jako argument dali pole t_pole, outputem funkce je pak pole
x_pole = x_sour_teorie(t_pole, t_0, x_0, v_0, alpha)

# pole souřadnice y
y_pole = y_sour_teorie(t_pole, t_0, y_0, v_0, alpha, g)

# Nastavíme velikost obrázku na 10 x 5 palců
plt.figure(figsize=(10, 5))

# Vykreslíme data do grafu
# První argument jsou x-ové souřadnice, druhý argument jsou y-ové souřadnice
# Třetí argument určuje podobu křivky - hvězdičky spojené přerušovanými čarami
# 3. argument je vlastně spojení dvou argumentů: marker='*', linestyle='-'
# Čtvrtý argument je pojmenování křivky (label)
plt.plot(x_pole, y_pole, "*-", label="Trajektorie míčku")

# Přidáme horizontální osu (osa y = 0) černou barvou
plt.axhline(y=0, color='k', linestyle='--')

# Přidáme vertikální osu (osa x = 0) černou barvou
plt.axvline(x=0, color='k', linestyle='--')

# Pojmenujeme osu x
plt.xlabel("x [m]")

# Pojmenujeme osu y
plt.ylabel("y [m]")

# Přidáme název grafu
plt.title("Vrh šikmý bez odporu prostředí")

# Přidáme legendu, aby bylo jasné, co jednotlivé křivky znamenají
plt.legend()

# Zobrazíme mřížku
#plt.grid()

# Zobrazíme graf
plt.show()
```

```python
# Grafy x(t) a y(t)

# Vytvoříme figure a dvě axes (podgrafy) vedle sebe
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# První graf (levý)
ax[0].plot(t_pole, x_pole, "*-", label="x(t)")
ax[0].set_xlabel("t [s]")
ax[0].set_ylabel("x [m]")
ax[0].set_title("x - souřadnice")
ax[0].legend(loc = "upper right")
ax[0].grid()

# druhý graf (pravý)
ax[1].plot(t_pole, y_pole, "*-", label="y(t)")
ax[1].set_xlabel("t [s]")
ax[1].set_ylabel("y [m]")
ax[1].set_title("y - souřadnice")
ax[1].legend(loc = "upper right")
ax[1].grid()

# Zobrazíme grafy
plt.tight_layout()  # Upraví rozložení, aby se grafy nepřekrývaly
plt.show()
```

## B Vybrané fyzikální příklady

### Úloha B1 - Vrh šikmý bez odporu prostředí (po druhé)

Pro teď se vrátíme k úvodnímu příkladu. Budeme ale předpokládat, že funkce  $x(t)$, $y(t)$ neznáme.

*Otázka:*

Je možné určit souřadnice míčku, aniž bychom potřebovali vědět jejich přesné vzorce?

*Odpověď:*

Ano, ze znalosti počátečních podmínek lze nalézt přibližnou polohu míčku v čase. Řešení není dáno ve formě jednoho vzorce, ale sekvence konkrétních čísel, které reprezenují polohu míčku v obou osách (mluvíme o numerickém řešení). Výchozím bodem přitom pro nás budou přibližné vzorce pro rychlost a zrychlení.

*Rychlost a zrychlení pomocí diferencí*

Jak zrychlení tak rychlost lze zapsat pomocí diferencí (rozdílů):
\begin{align}
a^x_i &= \frac{\Delta v^x_i }{\Delta t} = \frac{v^x_{i+1} - v^x_{i}}{t_{i+1} - t_i}\,,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, a^y_i = \frac{\Delta v^y_i }{\Delta t} = \frac{v^y_{i+1} - v^y_{i}}{t_{i+1} - t_i}\,, \\
v^x_i &= \frac{\Delta x_i }{\Delta t} = \frac{x_{i+1} - x_{i}}{t_{i+1} - t_i}\,,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, v^y_i = \frac{\Delta y_i }{\Delta t} = \frac{y_{i+1} - y_{i}}{t_{i+1} - t_i}\,.
\end{align}
Index $i$ přitom znamená, že se daná veličina počítá v čase $t_i$. Těchto časů budeme mít $N$ a indexovat budeme od nuly, tj. $i = 0,\, \dots\,, N-1$.

V ose $x$ žádná síla napůsobí a není tam tak žádné zrychlení, tzn.

 $$a^x = 0 \,\longrightarrow\, v^x_{0} = v^x_{1} = \dots = v^x_{N-1}\,,$$

přitom $v^x_{0}= v_0 \cos (\alpha)$.

V ose $y$ je zrychlení určeno tíhovým zrychlením, což znamená, že pro různá $i$ budou obecně různé rychlosti:

  $$a^y = -g \,\longrightarrow\, v^y_{i+1} = v^y_{i} - g \Delta t\,,$$

kde $v^y_{0} = v_0 \sin (\alpha)$,

Dostáváme tak následující schéma výpočtu:

* $x$-ová souřadnice

\begin{align}
x_{i+1} &= x_i + v_0 \cos (\alpha) \Delta t\,,
\end{align}

kde $x_0 = 0\,$,

* $y$-ová souřadnice

\begin{align}
y_{i+1} &= y_i + v^y_{i} \Delta t\,, \\
\end{align}
kde počáteční podmínky jsou $v^y_{0} = v_0 \sin (\alpha)$, $y_0 = 1\,$ a musíme, narozdíl od $x$-ové souřadnice, aktualizovat rychlost $v^y_{i+1} = v^y_{i} - g \Delta t\,$.



**Úkol**

Vytvořte funkce, které spočítají všechny okamžité polohy $(x_i)_{i=0}^{N-1}$ a $(y_i)_{i=0}^{N-1}$ pro zadaný časový krok $\Delta t$ a počet bodů $N$. Dalšími argumenty funkcí by pak měly být $x_0$, $y_0$, $v_0$, $\alpha$, $g$.

```python
# funkce vracící pole x -ových souřadnic
def x_sour_grid(N, dt, alpha, v_0, x_0):

  # vytvoříme list, který budeme plnit x -ovými souřadnicemi, jak je budeme postupně počítat; začneme s x_0 (počáteční podmínka)
  x_list = [x_0]

  # v každé iteraci spočítáme novou souřadnici a uložíme ji do listu
  # začínáme indexovat od 1 a končíme s N-1 (i = 0 už máme v listu uložený)
  for i in range(1, N):
    x_new = x_list[i-1] + v_0 * np.cos(alpha) * dt
    x_list.append(x_new)

  # není to nezbytné, ale i tak přetypujeme list a uložíme jako nové pole
  x_array = np.array(x_list)
  return x_array

# analogie pro y, jen musíme průběžně aktualizovat rychlost
def y_sour_grid(N, dt, alpha, v_0, y_0):

  y_list = [y_0]

  # počáteční rychlost v y-ové ose
  v = v_0 * np.sin(alpha)

  for i in range(1, N):
    y_new = y_list[i-1] + v * dt
    y_list.append(y_new)
    # aktualizace rychlosti
    v = v - g * dt

  y_array = np.array(y_list)
  return y_array
```

```python
# popřípadě můžeme opět napsat jedinou funkci, která vrátí x- i y- pole souřadnic
def xy_sour_grid(N, dt, alpha, v_0, x_0, y_0):

  x_list, y_list = [x_0], [y_0]
  v = v_0 * np.sin(alpha)

  for i in range(1, N):

    x_new = x_list[i-1] + v_0 * np.cos(alpha) * dt
    y_new = y_list[i-1] + v * dt

    x_list.append(x_new)
    y_list.append(y_new)

    v = v - g * dt

  x_array = np.array(x_list)
  y_array = np.array(y_list)
  return x_array, y_array
```

### Odbočka: Numerický integrátor vbudovaný přímo v Pythonu

Do Pythonu můžeme importovat knihovnu **scipy**, přesněji postačí jen její modul **integrate**, který má funkci odeint - je to funkce, jejíž pomocí můžeme najít numerické řešení diferenciální rovnice pro zadané počáteční podmínky.

```python
from scipy.integrate import odeint
```

Abychom s pomocí odeint dokázali najít $x$ a $y$ souřadnice, musíme pro ně umět napsat evoluční rovnice v podobě obecných diferenciálních rovnic (ODR):
$$
\frac{d x}{dt}\,,\,\,\,\,\,\frac{d y}{dt}
$$
To jsou rychlosti, pro něž bychom už potřebovali znát vzorce, jak je počítat, abychom mohli rovnou použít odeint k nalezení $x,y$. Vzorce pro rychlosti bohužel nemáme, ale pro jejich derivace, zrychlení, už naštěstí ano: $a_x = 0$, $a_y = -g$. Myšlenka je tak rozšířit úlohu z hledání $x,y$ na $x,y, v_x, v_y$.


Náš integrátor tak bude muset integrovat tyto 4 derivace:
$$
\frac{d x}{dt} = v_x\,,\,\,\,\,\,\frac{d y}{dt} = v_y,\,,\,\,\,\,\,\frac{d v_x}{dt}\,,\,\,\,\,\,\frac{d v_y}{dt}\,.
$$

Hledáme tedy vektor $\vec{S}(t) = (x(t), y(t), v_{x}(t), v_{y}(t))$ a k jeho nalezení potřebujeme znát jeho derivaci $d\vec{S}(t)/dt$ a počáteční podmínky.

```python
def dSdt(S, t, g):

  # aktuální hodnoty z vektoru S
  x, y, v_x, v_y = S

  # hodnoty derivací
  return [v_x, v_y, 0, -g]
```

```python
sol = odeint(dSdt, y0=[x_0, y_0, v_0 * np.cos(alpha), v_0 * np.sin(alpha)], t = t_pole_real, args = (g,))

print( sol.shape )
print( t_pole_real.shape )
```

### Úloha B2 - Vrh šikmý se započítáním odporu prostředí

Vrátíme se k předchozí úloze a zahrneme navíc odpor prostředí.


Tíhová síla není ve skutečnosti jediná síla, která na letící míček působí. Míček navíc pociťuje odporovou sílu prostředí, v jakém se pohybuje (v atmosféře). Tato síla je dána vztahem:

$$
\vec{F} = -\frac{1}{2} C \rho S v \vec{v}
$$

kde $C$ je činitel odporu (konstanta závislá na tvaru tělesa, více viz [zde](https://cs.wikipedia.org/wiki/%C4%8Cinitel_odporu)), $\rho$ je hustota vzduchu, $A$ je obsah příčného průřezu tělesem (pro míček o poloměru $r$ je to $A = \pi r^2$) pohybujícího se ve směru rychlosti o velikosti $v$.

Zrychlení tělesa souvisí s působící silou podle vztahu:

$$
\vec{a} = \frac{\vec{F}}{m}\,,
$$

z čehož pro jednotlivé komponenty $x$, $y$ plyne:

\begin{align}
a^x_i &= -\frac{1}{2m} C \rho A v_i v^x_i \\
a^y_i &= - g -\frac{1}{2m} C \rho A v_i v^y_i \\
\end{align}
kde $v_i = \sqrt{(v^x_i)^2 + (v^y_i)^2}$

Budeme uvažovat následující číselné hodnoty:

* $C = 0,47$, $\rho = 1,25~\mathrm{kg/m^3}$, $r = 3,2~\mathrm{cm}$, $m = 58~\mathrm{g}$

```python
C = 0.47
rho = 1.25
r = 0.032
m = 0.058
A = np.pi * r**2
```

**Úkol**

Vytvořte funkce, které spočítají všechny okamžité polohy $(x_i)_{i=0}^{N-1}$ a $(y_i)_{i=0}^{N-1}$ pro zadaný časový krok $\Delta t$ a počet bodů $N$. Dalšími argumenty funkcí by pak měly být $x_0$, $y_0$, $v_0$, $\alpha$, $g$, $m$, $C$, $\rho$.

```python
# Funkce počítající zrychlení v ose x pro aktuální hodnotu rychlosti
def a_xi(C, rho, A, v_xi, v_yi, m):
  v_i = np.sqrt(v_xi**2 + v_yi**2)
  return -0.5 * C * rho * A * v_i * v_xi / m

# Funkce počítající zrychlení v ose y pro aktuální hodnotu rychlosti
def a_yi(C, rho, A, v_xi, v_yi, m, g):
  v_i = np.sqrt(v_xi**2 + v_yi**2)
  return -g - 0.5 * C * rho * A * v_i * v_yi / m

# Funkce, která spočítá pole x-ové a y-ové souřadnice pro celý sledovaný časový interval
def xy_sour_u2(N, dt, x_0, y_0, v_0, alpha, g, A):

  # do těchto listů budeme ukládat oba souřadnice
  x_list, y_list = [x_0],  [y_0]

  # počáteční rychlost
  v_xi = v_0 * np.cos(alpha)
  v_yi = v_0 * np.sin(alpha)

  # iterujeme od 1 do N-1
  for i in range(1, N):

    # aktualizace souřadnic
    x_new = x_list[i-1] + v_xi * dt
    y_new = y_list[i-1] + v_yi * dt

    # uložení nových souřadnic do listů
    x_list.append(x_new)
    y_list.append(y_new)

    # aktualizace rychlostí
    v_xi, v_yi = v_xi + a_xi(C, rho, A, v_xi, v_yi, m) * dt,  v_yi + a_yi(C, rho, A, v_xi, v_yi, m, g) * dt

  # namísto s listy budeme pracovat s poli
  x_array = np.array(x_list)
  y_array = np.array(y_list)

  return x_array, y_array
```

```python
# Zadáme časový krok a počet bodů, v nichž budeme chtít provést vyčíslení
dt = 0.01
N = 400

# Sledovaná doba
T_obs = (N-1) * dt
print("Sledovaná doba = " + str(T_obs))

# Vytvoříme pole času - od t_0 do T_obs
t_pole = np.linspace(t_0, T_obs, N)

# Vytvoříme pole souřadnic x a y
x_pole_mkp_u2, y_pole_mkp_u2 = xy_sour_u2(N, dt, x_0, y_0, v_0, alpha, g, A)

plt.figure(figsize=(10, 5))
plt.plot(x_pole_mkp_u2, y_pole_mkp_u2, "*-", color = "red", label="Spočítaná trajektorie")
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.legend()
plt.show()
```

Nakonec ještě najdeme řešení s použitím metody odeint. Konkrétně je naším cílem najít vektor

$$
\vec{S}(t) = (x(t), y(t), v_{x}(t), v_{y}(t))\,,
$$

což uděláme tak, že zadáme vektor $d\vec{S}/dt$ a počáteční podmínky.

```python
def dSdt(S, t, C, rho, A, m, g):

  # aktuální hodnoty z vektoru S
  x, y, v_x, v_y = S

  v = np.sqrt(v_x**2 + v_y**2)

  # hodnoty derivací
  return [
      v_x,
      v_y,
      -0.5 * C * rho * A * v * v_x / m,
      -g - 0.5 * C * rho * A * v * v_y / m
      ]
```
